# pre3 16_1~20_1 concat

```python
import pandas as pd
```

```python
age16_1 = pd.read_csv('data/pre3_age16_1.csv')
age17_1 = pd.read_csv('data/pre3_age17_1.csv')
age18_1 = pd.read_csv('data/pre3_age18_1.csv')
age19_1 = pd.read_csv('data/pre3_age19_1.csv')
age20_1 = pd.read_csv('data/pre3_age20_1.csv')

age16_2 = pd.read_csv('data/pre3_age16_2.csv')
age17_2 = pd.read_csv('data/pre3_age17_2.csv')
age18_2 = pd.read_csv('data/pre3_age18_2.csv')
age19_2 = pd.read_csv('data/pre3_age19_2.csv')
age20_2 = pd.read_csv('data/pre3_age20_2.csv')
```

```python
# 냅다 합쳐버리기
age_concat1 = pd.concat([age16_1, age17_1, age18_1, age19_1, age20_1], ignore_index=True)
age_concat2 = pd.concat([age16_2, age17_2, age18_2, age19_2, age20_2], ignore_index=True)
```

16_2 ~ 20_2 의 경우 컬럼이 모두 29개이고 컬럼명도 일치함.

16_1 ~ 20_1 의 경우 각각의 컬럼들이 몇년도에 있었는지 확인할 필요가 있음

```python
# 기록할 DF 생성
## col_name 은 컬럼 이름, 
## 16~20 에는 1이면 있음, 0이면 없음, 
## total은 합계(몇개년도에 있었는지)

col_year_base = pd.DataFrame(columns=['col_name', 16, 17, 18, 19, 20, 'total'])
col_year_base
```

```python
# 16_1 중복처리 할 필요 없이 입력
age16_1_year = pd.DataFrame({'col_name' : age16_1.columns, 16:[1] * len(age16_1.columns), \
    17:[0] * len(age16_1.columns), 18:[0] * len(age16_1.columns), 19:[0] * len(age16_1.columns), 20:[0] * len(age16_1.columns)})
```

```python
col_year = pd.concat([col_year_base, age16_1_year])
col_year
```

```python
age17_1.columns
```

```python
# 17_1 있으면 값 수정, 없으면 행 추가
for i in age17_1.columns:
    if i in list(col_year['col_name']): # col_name에 있으면 True, 없으면 False(else 실행)
        index_i = list(col_year['col_name']).index(i)
        col_year.iloc[index_i, 2]=1
    else:
        add_i = pd.DataFrame({'col_name':[i], 16:0, 17:1, 18:0, 19:0, 20:0})
        col_year = col_year.append(add_i)
col_year
```

```python
# 18_1 있으면 값 수정, 없으면 행 추가
for i in age18_1.columns:
    if i in list(col_year['col_name']): # col_name에 있으면 True, 없으면 False(else 실행)
        index_i = list(col_year['col_name']).index(i)
        col_year.iloc[index_i, 3]=1
    else:
        add_i = pd.DataFrame({'col_name':[i], 16:0, 17:0, 18:1, 19:0, 20:0})
        col_year = col_year.append(add_i)
col_year
```

```python
# 19_1 있으면 값 수정, 없으면 행 추가
for i in age19_1.columns:
    if i in list(col_year['col_name']): # col_name에 있으면 True, 없으면 False(else 실행)
        index_i = list(col_year['col_name']).index(i)
        col_year.iloc[index_i, 4]=1
    else:
        add_i = pd.DataFrame({'col_name':[i], 16:0, 17:0, 18:0, 19:1, 20:0})
        col_year = col_year.append(add_i)
col_year
```

```python
# 20_1 있으면 값 수정, 없으면 행 추가
for i in age20_1.columns:
    if i in list(col_year['col_name']): # col_name에 있으면 True, 없으면 False(else 실행)
        index_i = list(col_year['col_name']).index(i)
        col_year.iloc[index_i, 5]=1
    else:
        add_i = pd.DataFrame({'col_name':[i], 16:0, 17:0, 18:0, 19:0, 20:1})
        col_year = col_year.append(add_i)
col_year
```

```python
# total 구하기
for i in range(len(col_year)):
    yr16 = col_year.iloc[i, 1]
    yr17 = col_year.iloc[i, 2]
    yr18 = col_year.iloc[i, 3]
    yr19 = col_year.iloc[i, 4]
    yr20 = col_year.iloc[i, 5]
    col_year.iloc[i, 6] = yr16+yr17+yr18+yr19+yr20
col_year
```

```python
# total 기준 정렬본
col_year_final = col_year.sort_values('total', ascending=False).reset_index()[['col_name', 16, 17, 18, 19, 20, 'total']]
col_year_final
```
