```python
import pandas as pd
```

```python
df7 = pd.read_csv('df7_code.csv',index_col = 0)
df8 = pd.read_csv('df8_code.csv',index_col = 0)
```

```python
len(df7)
```

```python
upper_7 = []
for i in df7.loc[:,'변수명']:
    upper_7.append(i.upper())
    
print(upper_7)
```

```python
upper8 = [i.upper() for i in df8.loc[:,'변수명']]
print(upper8)
```

```python
list8 = []
list7 = []
```

```python
for i in range(0,len(upper8)):
    for j in range(0,len(upper_7)):
        if upper8[i] == upper_7[j]:
            list8.append(i)
            list7.append(j)
            
```

```python
df8_2 = df8.iloc[list8]
df8_2
```

```python
df7_2 = df7.iloc[list7]
df7_2 
```

```python
df8_2.rename(columns={'변수설명':'변수설명8'}, inplace=True)
df7_2.rename(columns={'변수설명':'변수설명7'}, inplace=True)
df8_2
```

```python
lis = []
etc_lis = []
na_lis = []
for i in range(0,len(dd)):
    if dd.iloc[i].isnull().values.any():
        na_lis.append(i)
    else:
        if dd.iloc[i][1][0:2] in dd.iloc[i][2] or dd.iloc[i][2][0:2] in dd.iloc[i][1]:
            lis.append(i)
        else:
            etc_lis.append(i)
                
```

```python
dd_same =  dd.iloc[lis]
```

```python
na_dd = dd.iloc[na_lis]
```
