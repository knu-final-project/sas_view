```python
# 데이터 불러오기

import pandas as pd

# 7차 csv 파일 불러오기
df7 = pd.read_csv('df7_code.csv')
df7.head(5)
```

```python
# 8차 csv 파일 불러오기

df8 = pd.read_csv('df8_code.csv')
df8.head(5)
```

```python
# 7차 [변수명] 요소 대문자로 변경 ( 1446 row )

for i in range(len(df7)) :
    df7['변수명'][i] = df7['변수명'][i].upper()
print(df7)
```

```python
# 8차 [변수명] 요소 대문자로 변경 ( 905 row )

for i in range(len(df8)) :
    df8['변수명'][i] = df8['변수명'][i].upper()
print(df8)
```

```python
# 7차, 8차 concat

df78_con = pd.concat([df7, df8])
print(df78_con)
```

```python
# ['Unnamed: 0'] 컬럼 drop

df78_con = df78_con.drop(['Unnamed: 0'], axis = 1)
df78_con
```

```python
# index 리셋

df78_con = df78_con.reset_index(drop=True)
df78_con
```

```python
## 데이터 프레임 csv 파일로 저장

df78_con.to_csv('df78_con.csv', encoding='euc-kr')
```
