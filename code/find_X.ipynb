# 통합변수명 (정제 전) 기준으로 16_1~20_2 에는 있지만 통합변수명 에는 없는 컬럼 찾기

```python
# 7차/8차 변수명 정리 불러오기
col_7 = pd.read_csv('ns7_code.csv', encoding='cp949')
col_8 = pd.read_csv('ns8_code.csv', encoding='cp949')
col_7
```

```python
col_8
```

```python
col_7['변수명']
```

```python
# 7차 변수명 UPPER
for i in range(len(col_7['변수명'])):
    col_7['변수명'][i] = col_7['변수명'][i].upper()
col_7 = col_7[['변수명', '변수설명']]
col_7
```

```python
# 8차 변수명 UPPER
for i in range(len(col_8['변수명'])):
    col_8['변수명'][i] = col_8['변수명'][i].upper()
col_8 = col_8[['변수명', '변수설명']]
col_8
```

```python
# 7차 + 8차
col_concat = pd.concat([col_7, col_8])
col_concat
```

```python
col_concat = col_concat.drop_duplicates(['변수명'])
col_concat
```

```python
col_concat = col_concat.reset_index()[['변수명', '변수설명']]
col_concat
```

```python
# 변수정의서에는 없지만 df에 만 존재하는 컬럼을 추출하는 함수
def find_solo(df):

    # df에서 컬럼 추출
    df_col = df.columns

    # 컬럼 대문자화
    df_upcol = []
    for i in df_col:
        df_upcol.append(i.upper())

    # duo 둘다 있음 / solo : df에만 있음
    df_duo = []
    df_solo = []
    for i in df_upcol:
        if i in list(col_concat['변수명']): # col_concat 이 정의된 상태여야함
            df_duo.append(i)
        else:
            df_solo.append(i)

    return df_solo
```

```python
age16_1 = pd.read_csv('../data/pre2_age16_1.csv', index_col=False)
age17_1 = pd.read_csv('../data/pre2_age17_1.csv', index_col=False)
age18_1 = pd.read_csv('../data/pre2_age18_1.csv', index_col=False)
age19_1 = pd.read_csv('../data/pre2_age19_1.csv', index_col=False)
age20_1 = pd.read_csv('../data/pre2_age20_1.csv', index_col=False)

age16_2 = pd.read_csv('../data/pre2_age16_2.csv', index_col=False)
age17_2 = pd.read_csv('../data/pre2_age17_2.csv', index_col=False)
age18_2 = pd.read_csv('../data/pre2_age18_2.csv', index_col=False)
age19_2 = pd.read_csv('../data/pre2_age19_2.csv', index_col=False)
age20_2 = pd.read_csv('../data/pre2_age20_2.csv', index_col=False)
```

```python
age16_1_solo = find_solo(age16_1)
age17_1_solo = find_solo(age17_1)
age18_1_solo = find_solo(age18_1)
age19_1_solo = find_solo(age19_1)
age20_1_solo = find_solo(age20_1)

age16_2_solo = find_solo(age16_2)
age17_2_solo = find_solo(age17_2)
age18_2_solo = find_solo(age18_2)
age19_2_solo = find_solo(age19_2)
age20_2_solo = find_solo(age20_2)
```
