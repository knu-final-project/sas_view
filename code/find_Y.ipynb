```python
# Upper & 공백 제거
for i in range(len(dd_na['변수명'])):
    dd_na['변수명'][i] = dd_na['변수명'][i].upper().strip()
dd_na
```

```python
age16_1 = pd.read_csv('../data/pre2_age16_1.csv')
age17_1 = pd.read_csv('../data/pre2_age17_1.csv')
age18_1 = pd.read_csv('../data/pre2_age18_1.csv')
age19_1 = pd.read_csv('../data/pre2_age19_1.csv')
age20_1 = pd.read_csv('../data/pre2_age20_1.csv')

age16_2 = pd.read_csv('../data/pre2_age16_2.csv')
age17_2 = pd.read_csv('../data/pre2_age17_2.csv')
age18_2 = pd.read_csv('../data/pre2_age18_2.csv')
age19_2 = pd.read_csv('../data/pre2_age19_2.csv')
age20_2 = pd.read_csv('../data/pre2_age20_2.csv')
```

```python
def upper_col(df):
    df_col = df.columns
    up_col = []
    for i in df_col:
        up_col.append(i.upper().strip())
    return up_col
```

```python
all_col = upper_col(age16_1) + upper_col(age17_1) + upper_col(age18_1) + upper_col(age19_1) + upper_col(age20_1) + \
    upper_col(age16_2) + upper_col(age17_2) + upper_col(age18_2) + upper_col(age19_2) + upper_col(age20_2)
all_col
```

```python
len(set(all_col))
```

```python
dd_na_col = dd_na['변수명'].tolist()
len(dd_na_col)
```

```python
duo = []
solo = []
for i in dd_na_col:
    if i in upper_col(age18_1):
        duo.append(i)
    else:
        solo.append(i)
        
print(len(solo))
print(len(duo))
if 'BP16_1' in solo:
    print('없다')
else:
    print('있다')
```

```python
duo = []
solo = []
for i in dd_na_col:
    if i in set(all_col):
        duo.append(i)
    else:
        solo.append(i)
        
print(len(solo))
print(len(duo))
'HE_DM_HBA1C' in solo
```
