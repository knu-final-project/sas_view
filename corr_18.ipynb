```python
# 원본 복사 (index리셋)
pre_age18_2 = age18_2.reset_index().drop(['index'], axis=1)
pre_age18_2
```

8, 9 제거

```python
for i in age18_2_column[0:10]:
    col_index = list(pre_age18_2.columns).index(i)

    # 8 제거
    pre_age18_2_index = pre_age18_2[pre_age18_2[i] == 8][i].index

    for j in list(pre_age18_2_index):
        pre_age18_2.iloc[j,col_index] = np.nan
    
    # 9 제거
    pre_age18_2_index = pre_age18_2[pre_age18_2[i] == 9][i].index

    for j in list(pre_age18_2_index):
        pre_age18_2.iloc[j,col_index] = np.nan

pre_age18_2
```
