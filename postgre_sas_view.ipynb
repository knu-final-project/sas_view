pd_sas table 만들기 (pd_sas2 테이블 아님)

```python
# pd_sas 테이블 정의를 위해 컬럼명 List 추출
pd_sas_columns=pd_sas.columns
pd_sas_columns
```

```python
# 테이블 만드는 쿼리문 만들기 (Long)
text_query = 'CREATE TABLE IF NOT EXISTS pd_sas ('
for i in pd_sas_columns:
    text_query = text_query + i + ' varchar(50),'
text_query = text_query.rstrip(',') + ');'

text_query
```

```python
# postgre DB 연결
try:
    conn = psycopg2.connect(host="172.20.62.185",database="pgdb_final4", user="postgres", password="admin",port=5432)
    print('===접속 성공===')
    cur = conn.cursor()
except psycopg2.DatabaseError as db_err:
    print('접속 오류 !!')
    print(db_err)
```

```python
# pd_sas 테이블 CREATE
cur = conn.cursor()
try:
    cur.execute(text_query)
    conn.commit()
    print('success')
except:
    print('fail')
```
