# 데이터 컬럼명 대문자화 + 년도별 컬럼명 다른 것 통일

```python
import pandas as pd
```

```python
age16_1 = pd.read_csv('data/pre2_age16_1.csv')
age17_1 = pd.read_csv('data/pre2_age17_1.csv')
age18_1 = pd.read_csv('data/pre2_age18_1.csv')
age19_1 = pd.read_csv('data/pre2_age19_1.csv')
age20_1 = pd.read_csv('data/pre2_age20_1.csv')

age16_2 = pd.read_csv('data/pre2_age16_2.csv')
age17_2 = pd.read_csv('data/pre2_age17_2.csv')
age18_2 = pd.read_csv('data/pre2_age18_2.csv')
age19_2 = pd.read_csv('data/pre2_age19_2.csv')
age20_2 = pd.read_csv('data/pre2_age20_2.csv')

age16_1
```

```python
# 컬럼명을 대문자로 바꾸는 함수

def col_upper(df):
    df_col = df.columns
    df_upper = []
    for i in df_col:
        df_upper.append(i.upper().strip())
    
    df.columns = df_upper
```

```python
col_upper(age16_1)
col_upper(age17_1)
col_upper(age18_1)
col_upper(age19_1)
col_upper(age20_1)

col_upper(age16_2)
col_upper(age17_2)
col_upper(age18_2)
col_upper(age19_2)
col_upper(age20_2)
```

```python
# 컬럼 바꾸기! HE_DM_HBA1C
def change_col(df):
    # HE_DM_HBA1C -> HE_DM
    if 'HE_DM_HBA1C' in df.columns:
        df.rename(columns={'HE_DM_HBA1C':'HE_DM'}, inplace=True)
        print("'HE_DM_HBA1C' in df : True")
        print("'HE_DM_HBA1C' -> 'HE_DM' 변경")
    else:
        print("'HE_DM_HBA1C' in df : False")
    
    # T_HL_MODERATE -> T_HL_MOD
    if 'T_HL_MODERATE' in df.columns:
        df.rename(columns={'T_HL_MODERATE':'T_HL_MOD'}, inplace=True)
        print("'T_HL_MODERATE' in df : True")
        print("'T_HL_MODERATE' -> 'T_HL_MOD' 변경")
    else:
        print("'T_HL_MODERATE' in df : False")
```

```python
change_col(age16_1)
change_col(age17_1)
change_col(age18_1)
change_col(age19_1)
change_col(age20_1)

change_col(age16_2)
change_col(age17_2)
change_col(age18_2)
change_col(age19_2)
change_col(age20_2)
```

```python
'TOTAL_SLP_WK' in age18_1.columns
```

```python
# BP16_1 -> TOTAL_SLP_WK 값 시간 -> 분
def change_SLP(df):
    # BP16_1 -> TOTAL_SLP_WK
    if 'BP16_1' in df.columns:
        df['BP16_1'] = df['BP16_1']*60
        df.rename(columns={'BP16_1':'TOTAL_SLP_WK'}, inplace=True)
        print('BP16_1 -> TOTAL_SLP_WK 변경')
    else:
        print('BP16_1 없음')
    
    # BP16_2 -> TOTAL_SLP_WD
    if 'BP16_2' in df.columns:
        df['BP16_2'] = df['BP16_2']*60
        df.rename(columns={'BP16_2':'TOTAL_SLP_WD'}, inplace=True)
        print('BP16_2 -> TOTAL_SLP_WD 변경')
    else:
        print('BP16_2 없음')
```

```python
# 확인용
print('BP16_1' in age19_1.columns)
print('TOTAL_SLP_WK' in age19_1.columns)
```

```python
change_SLP(age16_1)
change_SLP(age17_1)
change_SLP(age18_1)
change_SLP(age19_1)
change_SLP(age20_1)

change_SLP(age16_2)
change_SLP(age17_2)
change_SLP(age18_2)
change_SLP(age19_2)
change_SLP(age20_2)
```
