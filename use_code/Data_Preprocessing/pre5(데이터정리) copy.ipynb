### 역순처리가 필요한 변수 변경 작업 실시

```python
new = [] # 새로운 변수 생성
nn = 0
for n in list(var_df[var_df.loc[:,'변경여부'] == 1].index): ## 변경여부가 1인 변수들 (역순처리 필요한 변수)
    nn += 1
    print(nn, n, var_df.iloc[n][0]) ## 변수 확인을 위함
    m = int(var_df.iloc[n][6]) ## 변경len : 응답문항 수
    if m != 0:
        m_list = var_df.iloc[n][7].split(',') # 변경list : 리스트[0] -> 0 ---- 리스트[n] -> 0+n 
    
    ## 역순 처리
        for i in range(0,len(age_1)):
            for j in range(0,m): 
                if age_1.loc[:,var_df.iloc[n][0]][i] == int(m_list[j]):
                    #print('변경전: ', age_1.loc[:,var_df.iloc[n][0]][i])
                    age_1.loc[:,var_df.iloc[n][0]][i] = j
                    #print('변경후:',age_1.loc[:,var_df.iloc[n][0]][i])
    else:
       new.append(n) 
```
