```python
## 라이브러리
import pandas as pd
```

```python
## 데이터 불러오기
df = pd.read_csv('data/diccsv.csv',encoding='cp949',index_col=0) # index_col = 0 : 0번째 컬럼을 인덱스로 사용하여라 
col = pd.read_csv('C:/sas_view/data/real_data1_final.csv',encoding='cp949',index_col=0) # 변수사전 데이터
df
```

2. 데이터 확인 작업 (var_c.csv / real_data1.csv)
* var_c.csv (상관분석을 통해 상관계수가 0.4이상인 변수들만 존재)


    * 제거여부
        * 0 : 질병에 영향을 미칠 가능성이 있다고 판단되는 변수들
        * 1 : 상관계수는 0.4 이상이지만 영향을 주지 않을 것 같다고 판단되는 변수들
    #
    * 변경여부
        * 0 : 설문지 응답문항이 큰 수일수록 긍정일 경우 (ex. 0 : 없음 / 1 : 긍정)
        * 1 : 부정이 더 큰 수에 해당하는 경우 (ex. 0 : 매우 그렇다 ~ 5 : 매우 그렇지 않다)

```python
## 제거여부와 변경여부가 추가된 변수사전 csv 불러오기
var_df = pd.read_csv('var_c.csv',encoding='cp949')
var_df
```

```python
## 해당 변수의 데이터들 추출 (test로 DI1_2 변수 확인)
n = 2
age_1.loc[:,var_df.iloc[n][0]]
```

### 변경이 필요한 변수 변경 작업 실시

```python
for n in list(var_df[var_df.loc[:,'변경여부'] == 1].index): ## 변경여부가 1인 변수들 (역순처리 필요한 변수)
    # print(var_df.iloc[n][0]) ## 변수 확인을 위함
    m = int(var_df.iloc[n][5]) # 변경len : 응답문항에서 가장 큰 수를 뜻함 ex) 1 2 3 4 5 -> 5 / 0 1 -> 1
    
    ## 역순 처리
    for i in range(0,len(age_1)):
        if  m > 1:    
            for j in range(0,m): 
                if age_1.loc[:,var_df.iloc[n][0]][i] == m-j: ## ex) 5 -> 1 / 4 -> 2 / 3 -> 3 / 2 -> 4 / 1 -> 5
                    #print('변경전: ', test[i])
                    age_1.loc[:,var_df.iloc[n][0]][i] = j+1
                    #print('변경후:',test[i])
        elif m == 1:
            #print('변경전: ', test[i])
            age_1.loc[:,var_df.iloc[n][0]][i] = abs(age_1.loc[:,var_df.iloc[n][0]][i]-1) ## 1 -> 0 / 0 -> 1 
            #print('변경후:',age_1.loc[:,var_df.iloc[n][0]][i])
        
```
