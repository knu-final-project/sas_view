```python
import pandas as pd

# csv 파일 불러오기

df = pd.read_csv('C:/sas_view/use_code/Data_Preprocessing/data/reverse2_df.csv', index_col = 0)
df = df.drop(columns = ['Unnamed: 0.1'])
df
```

```python
var_c = pd.read_csv('C:/sas_view/use_code/var_c.csv', encoding = 'cp949')
var_c
```

```python
# var_c 에서 [제거여부] 컬럼이 0 인 것만 추출하여 리스트 생성

var_0 = var_c[var_c.loc[ : , '제거여부'] == 0]
var_0 # 335 

use_var = list(var_0.loc[ : , '변수명'])
len(use_var)
```

```python
# df 에서 use_var 컬럼에 속하는 컬럼만 추출 

reverse3_df = df.loc[ : , df.columns.isin(use_var)]
reverse3_df
```

```python
# csv 파일로 저장

reverse3_df.to_csv('C:/sas_view/use_code/Data_Preprocessing/data/reverse3_df.csv')
```
