```python
### 파일 불러오기
df7 = pd.read_csv('df7_code.csv',index_col = 0)
df8 = pd.read_csv('df8_code.csv',index_col = 0)
```

```python
### for문을 사용하여 7차 변수들 대문자로 변환
upper_7 = []
for i in df7.loc[:,'변수명']:
    upper_7.append(i.upper())
    
print(upper_7)
```

```python
## 요약for문 사용하여 8차 변수 대문자 변환
upper_8 = [i.upper() for i in df8.loc[:,'변수명']]
print(upper_8)
```

```python
## 7차 8차 서로 동일한 변수명만 리스트에 append
list8 = []
list7 = []
for i in range(0,len(upper_8)):
    for j in range(0,len(upper_7)):
        if upper_8[i] == upper_7[j]:
            list8.append(i)
            list7.append(j)
            
```

```python
### merge하기전 변수명 변경
df8.rename(columns={'변수설명':'변수설명8'}, inplace=True)
df7.rename(columns={'변수설명':'변수설명7'}, inplace=True)
df8
```

```python
df7['변수명'] = upper_7
df8['변수명'] = upper_8
```

```python
### 뽑힌 8차 변수들
df8_2 = df8.iloc[list8]
df8_2
```

```python
### 뽑힌 7차 변수들
df7_2 = df7.iloc[list7]
df7_2 
```

```python
print(len(list8),len(list7))
```

```python
### 같은 변수들만 merge한 데이터프레임
merge_df = pd.merge(df7_2,df8_2)
merge_df
```

```python
### 전체를 merge
m_df = pd.merge(df7,df8,how='outer')
```

```python
#### m_df : 전체
m_df
```

```python
### merge_df : 중복되는 행만
merge_df
```

```python
### 같은변수에서 설명이 같은 것들만 
merge_df[merge_df.iloc[:,1] == merge_df.iloc[:,2]]
```

```python
### 같은변수에서 설명이 다른 것들만 
merge_df[merge_df.iloc[:,1] != merge_df.iloc[:,2]]
```

```python
## 전체merge df에서 설명이 같은 변수들 뽑기
same = m_df[m_df.iloc[:,1] == m_df.iloc[:,2]]
same.to_csv('same_code.csv')
```

```python
## 전체merge df에서 설명이 다른 변수들 뽑기
dd = m_df[m_df.iloc[:,1] != m_df.iloc[:,2]]
```

```python
## dd 데이터프레임 중 나누어주기
lis = []
etc_lis = []
na_lis = []
for i in range(0,len(dd)):
    if dd.iloc[i].isnull().values.any():
        na_lis.append(i) ## Na 포함
    else:
        if dd.iloc[i][1][0:2] in dd.iloc[i][2] or dd.iloc[i][2][0:2] in dd.iloc[i][1]:
            lis.append(i) ## 변수설명이 같지만 미세하게 달라서 뽑힌 변수들
        else:
            etc_lis.append(i) ## 그 외 변수들 (아예 다른 변수)
                
```

```python
## dd 중 변수설명이 같은 변수
dd_same =  dd.iloc[lis]
```

```python
## Na값을 포함한 변수
na_dd = dd.iloc[na_lis]
```

```python
## 그 외 변수
dd.iloc[etc_lis]
```

```python
dd.iloc[etc_lis][2:] ## 설명이 똑같은 것으로 판별된 변수들
```

```python
## 7차 8차 서로 다른 변수들 모음
na_dd.to_csv('dd_na_code.csv',encoding='cp949') # NA 존재 파일
dd.iloc[etc_lis].to_csv('dd_code.csv',encoding='cp949')
```

```python
## 7차, 8차 같은 변수들
same2 = pd.concat([same,dd_same,dd.iloc[etc_lis][2:]]) 
same2
```

```python
## 7차 8차 변수 동일
same2.to_csv('same_code.csv')
```
